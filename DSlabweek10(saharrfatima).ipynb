```python
class SequentialSearch:
    def __init__(self, data):
        self.data = data

    def search(self, target):
        steps = 0
        for index in range(len(self.data)):
            steps += 1
            if self.data[index] == target:
                return index, steps
        return -1, steps


arr = [8, 15, 6, 42, 4, 33]
obj = SequentialSearch(arr)
tests = [15, 23, 33, 99, 8]   

for t in tests:
    result, steps = obj.search(t)
    if result != -1:
        print(f"Target {t} found at index {result} in {steps} steps.")
    else:
        print(f"Target {t} not found after {steps} steps.")
```

```python
class BinarySearch:
    def __init__(self, data):
        self.data = sorted(data)

    def search(self, target):
        low = 0
        high = len(self.data) - 1

        while low <= high:
            mid = (low + high) // 2

            if self.data[mid] == target:
                return mid
            elif target > self.data[mid]:
                low = mid + 1
            else:
                high = mid - 1

        return -1


arr = [15, 77, 23, 42, 4, 37]
obj = BinarySearch(arr)

print(obj.search(37))
print(obj.search(99))
```

```python
Time Complexity

Sequential Search:
→ O(n) (slow for large lists)
Binary Search:
→ O(log n) (much faster)
When to Use Sequential Search?
When data is not sorted
When the list is small
When you want a simple search method
When to Use Binary Search?
When data is sorted
When the list is large
When you need fast searching
Why Sorted Data Is Needed for Binary Search?
Because binary search decides to go left or right based on the middle value.
This is only possible if the list is sorted.
If the data is not sorted → you cannot know which side to search.
```

```python
import random
data = list(range(1, 100001))   
# Sequential Search
def sequential_search(arr, target):
    steps = 0
    for i in range(len(arr)):
        steps += 1
        if arr[i] == target:
            return i, steps
    return -1, steps
def binary_search(arr, target):
    low = 0
    high = len(arr) - 1
    steps = 0

    while low <= high:
        steps += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, steps
        elif target > arr[mid]:
            low = mid + 1
        else:
            high = mid - 1

    return -1, steps
target = random.choice(data)
print("Random Target:", target)

seq_index, seq_steps = sequential_search(data, target)

bin_index, bin_steps = binary_search(data, target)

print(f"Sequential Search: found at index {seq_index} in {seq_steps} steps.")
print(f"Binary Search: found at index {bin_index} in {bin_steps} steps.")
```

```python
 Why is Sequential Search slow for big lists?

Because it checks every element one by one.
If the list is very large, it may need to look through thousands or millions of items before finding the target.

2. Why does Binary Search require sorted data?

Binary Search decides to go left or right by comparing the target with the middle value.
This decision only makes sense if the data is sorted.

If the list is not sorted → you cannot know which side the target is on.

3. Explain low, high, and mid in Binary Search.

low → starting index of the current search area

high → ending index of the current search area

mid → the middle index between low and high

mid
=
low + high
2
mid=
2
low + high
	​


Binary Search checks the mid value and then adjusts low or high.

4. Can Binary Search be used for linked lists? Why?

No.
Binary Search needs direct access to the middle element using indexing.
Linked lists cannot jump to the middle; they must move node by node.
So binary search becomes slow and useless on linked lists.

5. Real-life examples
Sequential Search Example (Linear Search)

Looking for a name in an unsorted attendance sheet by reading from top to bottom.

Binary Search Example

Searching a word in a dictionary, which is alphabetically sorted:
You open near the middle, compare, and go left or right.
```
